```python
# 2. Create a dictionary of employee id and department.
{i["id"]:i["dept"] for i in employees}
```

```python
# 3. Filter only IT department employees and create a dictionary of name and salary.
{i["id"]:i["dept"] for i in employees if i['dept']=="IT"}
```

```python
# 10% bonuse to all hr 
{i["name"] : round(i["salary"] * 1.10) for i in employees if i['dept']=="HR" }
```

```python
# 4. Create a dictionary of employees earning more than 60,000 (id as key and salary as value).
{i["name"] : i["salary"] for i in employees if i["salary"]>60000}
```

```python
# 5. Create a dictionary of employees having more than 4 years of experience (name: experience).
{i["name"]:i["dept"] for i in employees if i["exp"]>4}
```

```python
# 6. Create a dictionary of city-wise employee count.
D={}
for i in employees:
    if i['city'] not in D:
        D[i['city']]=1
    else:
        D[i['city']]+=1
D
```

```python
# 7. Convert all employee names to uppercase and store them as id: NAME.
{emp["id"]: emp["name"].upper() for emp in employees}
```

```python
# 8. Add 10% salary hike to each employee and create a dictionary of name: updated salary.
updated_salary = {i["name"] : round(i["salary"] * 1.10) for i in employees}
print(updated_salary)
```

```python
# 9. Create a dictionary of department-wise total salary.
dept_salary = {dept: sum([emp["salary"] for emp in employees if emp["dept"] == dept])
for dept in {emp["dept"] for emp in employees}}
print(dept_salary)
```

```python
# 10. Create a dictionary marking employees as “Senior” or “Junior” based on experience (Senior > 5).
{emp["id"]: ("Senior" if emp["exp"] > 5 else "Junior") for emp in employees}
```

```python
# 11. Create a dictionary of all employees from Delhi (id: name).
dict({i["id"]:i["name"] for i in employees if i['city']=="Delhi"})
```

```python
# 12. Create a dictionary of employee names and the length of the name.
{i["name"]:len(i["name"]) for i in employees}
```

```python
# 13. Create a dictionary to find the highest salary in each department.
D = {}
for emp in employees:
    dept = emp["dept"]
    salary = emp["salary"]

    if dept not in D:
        D[dept] = salary
    else:
        D[dept] = max(D[dept], salary)
print(D)
```

```python
# Sir wala sum of salary in each department.
D=[(x["dept"],x['salary']) for x in employees]
D1={}
for i in D:
    if i[0] not in D1:
        D1[i[0]]=i[1]
    else:
        D1[i[0]]+=i[1]
D1
```
